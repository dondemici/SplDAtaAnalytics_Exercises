```python
import matplotlib.pyplot as plt

# Set up figure
fig, ax = plt.subplots(figsize=(14, 10))
ax.axis('off')
ax.set_title(" Data Analytics Concepts (Simplified Diagram)", fontsize=18, fontweight='bold')

# Define grouped concepts with short explanations
concept_groups = {
    " Center of Data": [
        ("Mean", "Average value"),
        ("Median", "Middle value"),
        ("Mode", "Most frequent value")
    ],
    " Spread / Variability": [
        ("Range", "Biggest - Smallest"),
        ("Standard Deviation", "Average distance from mean"),
        ("Variance", "Squared average distance"),
        ("IQR", "Middle 50% spread")
    ],
    " Shape of Distribution": [
        ("Skewness", "Leans left or right"),
        ("Kurtosis", "Peaked or flat?")
    ],
    " Standardized Scores": [
        ("Z-score", "How far from mean in SDs"),
        ("T-score", "Z-score for small data")
    ],
    " Relationships": [
        ("Covariance", "How 2 things change together"),
        ("Correlation", "How strongly 2 things move together (1 to 1)"),
        ("Regression", "Predict one thing from another"),
        ("R-squared", "How well the prediction fits")
    ],
    " Statistical Testing": [
        ("P-value", "Is it just luck?"),
        ("Chi-square", "Are categories related?"),
        ("ANOVA", "Compare 3+ groups")
    ],
    " Position & Estimation": [
        ("Percentile", "Your rank in the group"),
        ("Confidence Interval", "Range for true value")
    ]
}

# Plot each category in a box
x_start = 0.1
y_start = 0.9
y_step = 0.12

for i, (group, items) in enumerate(concept_groups.items()):
    y = y_start - i * y_step
    ax.text(x_start, y, group, fontsize=14, fontweight='bold', color="#2a558c")
    for j, (concept, desc) in enumerate(items):
        y_offset = y - (j + 1) * 0.035
        ax.text(x_start + 0.02, y_offset, f" {concept}: {desc}", fontsize=12)

plt.tight_layout()
plt.show()
```
